<u>tokenize and lemmatize text</u>

```python
from nlp.preprocessing import tokenize_and_lemmatize, en_stopwords

with open('assets/text/example1.txt', 'r', encoding='utf-8') as f:
    text = f.read()

tokens = tokenize_and_lemmatize(text, en_stopwords)
tokens = tokens.split()
print(tokens)
```

<u>pandas n gram analysis</u>

```python
from nltk.util import ngrams 
import pandas as pd
import matplotlib.pyplot as plt

unigrams = pd.Series(list(ngrams(tokens, 1))).value_counts()
unigrams[:10].sort_values().plot.barh(color="lightsalmon", width=0.8, figsize=(12,6))
plt.title('Top 10 Unigrams')
plt.xlabel('Frequency')
plt.show()
```

```python
# bigrams

bigrams = pd.Series(list(ngrams(tokens, 2))).value_counts()
bigrams[:12].sort_values().plot.barh(color="lightblue", width=0.8, figsize=(12,3))
plt.title('Top 10 Bigrams')
```
