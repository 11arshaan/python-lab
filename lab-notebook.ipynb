```python
from nlp.preprocessing import en_stopwords, remove_stopwords_simple, remove_stopwords_tokenized, remove_stopwords_preserve_punct

text = "The colors of the sky are beautiful today, aren't they? I hope the weather stays nice for the weekend!"

filtered_text_1 = remove_stopwords_simple(text, en_stopwords)
filtered_text_2 = remove_stopwords_tokenized(text, en_stopwords)
filtered_text_3 = remove_stopwords_preserve_punct(text, en_stopwords)

print("Original text:", text)
print("\nMethod 1 (Simple split):", filtered_text_1)
print("\nMethod 2 (Tokenized, alphanumeric only):", filtered_text_2)
print("\nMethod 3 (Tokenized, preserve punctuation):", filtered_text_3)
```
